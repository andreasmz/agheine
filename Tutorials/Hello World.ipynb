```python
import numpy as np
import matplotlib.pyplot as plt

X = range(0,10000)
Y = np.round(3*np.random.random(10000)-1.5, 0)
Y = np.cumsum(Y)
plt.plot(X,Y)
plt.title("Random Walk")
plt.show()
```
